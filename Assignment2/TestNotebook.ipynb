```python
def convert_cost_to_dict(cost):
    
    size = len(cost)
    graph = dict()
    for i in range(1, size):
        graph[i] = []
        for j in range(1, size):
            if cost[i][j] > 0:
                graph[i].append(j)

    return graph


def DFS_Traversal(cost, start_point, goals):
    l = []
    size = len(cost)
    graph = convert_cost_to_dict(cost)
    explored = set()
    
    
    def dfs(graph, start_point, goals):
        
        if start_point in explored:
            return []
        
        
        explored.add(start_point)
        l.append(start_point)
        
       
            
        
        if(start_point in goals):
            return l
        
        for i in graph[start_point]:
            if i not in explored:
                return dfs(graph, i, goals)
        if explored:
            explored.pop()
        if l:
            l.pop()
    
    dfs(graph, start_point, goals)
    return l


def UCS_Traversal(cost, start_point, goals):
    l = []
    size = len(cost)
    graph = convert_cost_to_dict(cost)
    frontier = []
    frontier.append((start_point, [start_point], 0))
    explored = set()
    
    while frontier != []:
        
        frontier = sorted(frontier, key = lambda x: x[2])
        current_node, current_path, current_cost = frontier[0]
        frontier = frontier[1:]

        if current_node in explored:
            continue
        else:
            explored.add(current_node)

        if current_node in goals:
            return current_path
        else:
            for i in graph[current_node]:
                if i not in explored:
                    frontier.append((i, current_path + [i], current_cost + cost[current_node][i]))
        

                    
    return l
                    
                    

def A_star_Traversal(cost, start_point, goals, heuristic):
    l = []
    size = len(cost)
    graph = convert_cost_to_dict(cost)
    frontier = []
    frontier.append((start_point, [start_point], 0))
    explored = set()
    
    while frontier != []:
        
        frontier = sorted(frontier, key = lambda x: x[2])
        current_node, current_path, current_cost = frontier[0]
        frontier = frontier[1:]
        
        if current_node in explored:
            continue
        else:
            explored.add(current_node)
        
        if current_node in goals:
            return current_path
        else:
            for i in graph[current_node]:
                if i not in explored:
                    frontier.append((i, current_path + [i], current_cost + cost[current_node][i] + heuristic[i] - heuristic[current_node]))

    return l



'''
Function tri_traversal - performs DFS, UCS and A* traversals and returns the path for each of these traversals 

n - Number of nodes in the graph
m - Number of goals ( Can be more than 1)
1<=m<=n
Cost - Cost matrix for the graph of size (n+1)x(n+1)
IMP : The 0th row and 0th column is not considered as the starting index is from 1 and not 0. 
Refer the sample test case to understand this better

Heuristic - Heuristic list for the graph of size 'n+1' 
IMP : Ignore 0th index as nodes start from index value of 1
Refer the sample test case to understand this better

start_point - single start node
goals - list of size 'm' containing 'm' goals to reach from start_point

Return : A list containing a list of all traversals [[],[],[]]
1<=m<=n
cost[n][n] , heuristic[n][n], start_point, goals[m]

NOTE : you are allowed to write other helper functions that you can call in the given fucntion
'''

def tri_traversal(cost, heuristic, start_point, goals):
    l = []


    t1 = DFS_Traversal(cost, start_point, goals)
    #send whatever parameters you require 

    t2 = UCS_Traversal(cost, start_point, goals)
    #send whatever parameters you require 

    t3 = A_star_Traversal(cost, start_point, goals, heuristic)
    #send whatever parameters you require 


    l.append(t1)
    l.append(t2)
    l.append(t3)
    return l


```
